```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sal = pd.read_csv('https://raw.githubusercontent.com/catprokhorova/HW_datasets/main/EDA_1/salary_dataset.csv')
sal.head()
```

# Основные статистики

```python
sal.info()
```

```python
sal.describe()
```

```python
# Настройка стиля для графиков
sns.set(style="whitegrid")

# Гистограммы для числовых переменных
sal.hist(bins=30, figsize=(15, 10), edgecolor='black')
plt.suptitle('Гистограммы числовых переменных', fontsize=16)
plt.show()
```

```python
# Box plot для выявления выбросов
plt.figure(figsize=(15, 10))
sns.boxplot(data=sal, orient='h', palette="Set2")
plt.title('Box Plot для выявления выбросов', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Years of Experience', y='Salary', data=sal, palette="Set2")
plt.title('Распределение зарплат в зависимости от уровня опыта', fontsize=16)
plt.xlabel('Уровень опыта')
plt.ylabel('Зарплата (USD)')
plt.show()
```

**Вывод: Наблюдаем, что с увеличением уровня опыта средняя зарплата растет. Это ожидаемо, так как более опытные сотрудники обычно зарабатывают больше.**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Education Level', y='Salary', data=sal, palette="Set2")
plt.title('Распределение зарплат в зависимости от уровня образования', fontsize=16)
plt.xlabel('Уровень образования')
plt.ylabel('Зарплата (USD)')
plt.show()
```

**Вывод: Видим, что сотрудники с высшим уровнем образования (например, магистры и доктора) получают более высокие зарплаты по сравнению с сотрудниками с бакалаврским образованием.**

```python
sal['Salary'] = pd.to_numeric(sal['Salary'], errors='coerce')

# Построение графика
plt.figure(figsize=(15, 10))
sns.boxplot(x='Country', y='Salary', data=sal, palette="Set2")
plt.title('Распределение заработной платы по странам', fontsize=16)
plt.xlabel('Страна')
plt.ylabel('Зарплата (USD)')
plt.xticks(rotation=90)
plt.show()
```

**Вывод: График показывает, как распределяются зарплаты среди различных стран. Вы можете увидеть медианные значения зарплат для каждой страны, а также диапазон значений и выбросы.Самая высокая заработная плата может быть в Канаде, а наиболее усредненные значения в Австралии.** 

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Race', y='Salary', data=sal, palette="Set2")
plt.title('Распределение заработной платы по расам', fontsize=16)
plt.xlabel('Раса')
plt.ylabel('Зарплата (USD)')
plt.xticks(rotation=45)
plt.show()
```

**Вывод: медианные показатели зарплат не зависят от расовой принадлежности работников, так как показатели у всех примерно равные исходя из графика. Тем не менее наименьшим показателем являютсяграждане Китая, что скорее всего связано с большим количеством населения.**
