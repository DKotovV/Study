```python
# Функция для поиска выбросов с использованием IQR
def find_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Поиск выбросов для каждого непрерывного столбца
continuous_columns = ['rectal_temperature', 'pulse', 'respiratory_rate']
outliers = {col: find_outliers(col) for col in continuous_columns}

for col, outlier_data in outliers.items():
    print(f"Выбросы в столбце {col}:")
    print(outlier_data)
```
